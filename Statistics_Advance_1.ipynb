## 1. What is a random variable in probability theory?

A **random variable (RV)** is a function that assigns a real number to each outcome in a sample space.



### Types of Random Variables:

1. **Discrete Random Variable**

   * Takes countable values (finite or countably infinite).
   * Example: Number of heads in 3 coin tosses → {0, 1, 2, 3}.

2. **Continuous Random Variable**

   * Takes values from an uncountable set (usually an interval of real numbers).
   * Example: Height of a student (can be any real value in a range).



### Example:

* Experiment: Roll a die.
* Sample Space: $S = \{1, 2, 3, 4, 5, 6\}$.
* Define random variable $X$ = "the number showing on the die."
  → Here, $X$ is a **discrete random variable** taking values {1, 2, 3, 4, 5, 6}.

Another example:

* Experiment: Measure the time (in seconds) it takes for a bulb to burn out.
* Random Variable $T$ = "lifetime of the bulb."
  → $T$ is a **continuous random variable**.





## 2. What are the types of random variables?


In **probability theory**, random variables are mainly classified into two types based on the nature of their possible values:


## **1. Discrete Random Variable**

* **Definition**: A random variable that can take a **finite or countably infinite set** of values.
* Values are **separate and distinct** (no fractions in between).
* Probability is described using a **Probability Mass Function (PMF)**.

**Examples:**

* Number of heads in 5 coin tosses → {0,1,2,3,4,5}.
* Score when rolling a die → {1,2,3,4,5,6}.
* Number of students present in a class.


## **2. Continuous Random Variable**

* **Definition**: A random variable that can take **any value within an interval** of real numbers (uncountably infinite).
* Values are not distinct, but spread over a continuum.
* Probability is described using a **Probability Density Function (PDF)**.
* The probability of a single exact value is **0**, so we calculate probabilities over intervals.

**Examples:**

* Height or weight of a person.
* Time taken to run a race.
* Temperature on a given day.


## **3. Mixed Random Variable** (sometimes considered a separate type)

* Some situations involve a mix of discrete and continuous values.

**Example:**

* Insurance claims:

  * Probability of **no claim (0)** (discrete).
  * Otherwise, claim amount can be any positive real number (continuous).



## 3.  Explain the difference between discrete and continuous distributions.


## **1. Discrete Distribution**

* **Definition:** Probability distribution of a **discrete random variable** (countable outcomes).
* Uses a **Probability Mass Function (PMF)** → assigns probability to each possible value.
* Total probability:

  $$
  \sum P(X=x_i) = 1
  $$

**Examples:**

* Binomial Distribution (number of heads in coin tosses).
* Poisson Distribution (number of phone calls in an hour).


## **2. Continuous Distribution**

* **Definition:** Probability distribution of a **continuous random variable** (uncountable outcomes over an interval).
* Uses a **Probability Density Function (PDF)** → probability is found **over an interval**, not at a single point.
* Total probability:

  $$
  \int_{-\infty}^{\infty} f(x) \, dx = 1
  $$
* Probability of any exact value (e.g., $P(X=2)$) is **0**.

**Examples:**

* Normal Distribution (heights of students).
* Exponential Distribution (time until a machine fails).



## 4. What is a binomial distribution, and how is it used in probability?


## **Binomial Distribution**

A **Binomial Distribution** is a **discrete probability distribution** that describes the number of **successes** in a fixed number of independent trials, where each trial has only **two possible outcomes**:

* Success (with probability $p$)
* Failure (with probability $q = 1 - p$)



### **Mathematical Formula**

The probability of getting exactly $k$ successes in $n$ trials is:

$$
P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}
$$

Where:

* $n$ = number of trials
* $k$ = number of successes
* $p$ = probability of success in a single trial
* $q = 1-p$ = probability of failure
* $\binom{n}{k} = \frac{n!}{k!(n-k)!}$ = number of ways to choose $k$ successes


### **Conditions for Binomial Distribution**

1. Fixed number of trials ($n$).
2. Each trial is independent.
3. Only two outcomes per trial (success/failure).
4. Probability of success ($p$) remains constant.



### **Example**

* Toss a coin 5 times.
* Probability of Head (success) = $p = 0.5$.
* What’s the probability of getting exactly 3 heads?

$$
P(X=3) = \binom{5}{3} (0.5)^3 (0.5)^2
= \frac{5!}{3!2!} (0.125)(0.25)
= 10 \times 0.03125
= 0.3125
$$

So, probability = **31.25%**.


### **Uses of Binomial Distribution**

1. **Quality control** – probability that a batch has certain defective items.
2. **Elections** – probability a candidate wins certain votes out of a sample.
3. **Sports** – probability a player scores certain goals in a fixed number of attempts.
4. **Medicine** – probability a patient responds to treatment in clinical trials.


## 5. What is the standard normal distribution, and why is it important?


## **Standard Normal Distribution**

The **standard normal distribution** is a special case of the **normal distribution**.

* It is a **bell-shaped curve** that is **symmetric around 0**.
* It has:

  * **Mean (μ) = 0**
  * **Standard deviation (σ) = 1**

The probability density function (PDF) is:

$$
f(z) = \frac{1}{\sqrt{2\pi}} e^{-\frac{z^2}{2}}, \quad -\infty < z < \infty
$$

Here, the variable is denoted as **Z** (called the *standard normal random variable*).


## **Why It’s Important**

1. **Simplifies Calculations**

   * Any normal random variable $X \sim N(\mu, \sigma^2)$ can be converted (standardized) into the standard normal variable:

   $$
   Z = \frac{X - \mu}{\sigma}
   $$

   This process is called **standardization**.

2. **Z-Scores**

   * A **Z-score** tells how many standard deviations an observation is from the mean.
   * Example: $Z=2$ means the value is 2σ above the mean.

3. **Probability Tables (Z-Table)**

   * We use the standard normal distribution to compute probabilities and areas under the curve.
   * Example: Probability that $Z < 1.96$ ≈ 0.975 (commonly used in hypothesis testing).

4. **Foundation for Statistics**

   * Many statistical methods (confidence intervals, hypothesis tests, regression analysis) are based on standard normal distribution.

## **Example**

Suppose $X \sim N(100, 15^2)$ (IQ scores with mean = 100, std dev = 15).
What’s the probability that a person has IQ less than 115?

Step 1: Convert to Z

$$
Z = \frac{X - \mu}{\sigma} = \frac{115 - 100}{15} = 1
$$

Step 2: Look up in Z-table:

$$
P(Z < 1) = 0.8413
$$

So, probability = **84.13%**.



## 6. What is the Central Limit Theorem (CLT), and why is it critical in statistics?


## **Central Limit Theorem (CLT)**

The **Central Limit Theorem (CLT)** states that:

> If we take a large number of random samples of size $n$ from any population with mean $\mu$ and standard deviation $\sigma$, then the **sampling distribution of the sample mean** will approximate a **normal distribution**, regardless of the original population’s distribution, provided $n$ is sufficiently large.

### **Key Points of CLT**

1. Works for **any population distribution** (normal, skewed, uniform, etc.).
2. The larger the sample size ($n$), the better the approximation (usually $n \geq 30$ is considered enough).
3. The **sampling distribution of the mean** has:

   * Mean = $\mu$ (same as population mean)
   * Standard deviation = $\frac{\sigma}{\sqrt{n}}$ (called the **standard error**)


### **Why is CLT Critical?**

1. **Foundation of Inferential Statistics**

   * Allows us to make conclusions about a population from a sample.

2. **Justifies Normal Approximations**

   * Even if the population is not normal, the sample mean distribution becomes nearly normal for large $n$.

3. **Basis for Confidence Intervals & Hypothesis Testing**

   * We can use **Z-scores** and **t-tests** thanks to CLT.

4. **Real-World Applications**

   * Quality control, polling, finance, medicine → whenever we take samples, CLT ensures reliability of results.


### **Example**

Suppose exam scores in a college are skewed (not normal) with:

* Population mean $\mu = 70$
* Standard deviation $\sigma = 10$

If we take samples of size $n = 50$:

* Sampling distribution of sample mean will be **approximately normal**.
* Mean of sample means = $70$.
* Standard error = $\frac{10}{\sqrt{50}} \approx 1.41$.

So, even though the original scores are skewed, the sample means follow a near-normal distribution.



## 7.  What is the significance of confidence intervals in statistical analysis?



## **Confidence Interval (CI)**

A **confidence interval** is a range of values, derived from sample data, that is likely to contain the **true population parameter** (like mean, proportion, etc.) with a certain level of confidence.


### **Formula (for mean with known σ):**

$$
CI = \bar{X} \pm Z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}
$$

Where:

* $\bar{X}$ = sample mean
* $Z_{\alpha/2}$ = critical value from the standard normal distribution (e.g., 1.96 for 95% confidence)
* $\sigma$ = population standard deviation
* $n$ = sample size

## **Significance of Confidence Intervals**

1. **Estimation of Population Parameters**

   * Instead of giving just a single estimate (sample mean), CI provides a **range of plausible values** for the population mean/proportion.

2. **Quantifies Uncertainty**

   * Shows the **margin of error** in our estimate. A wider CI = more uncertainty, narrower CI = more precision.

3. **Connection with Significance Testing**

   * If a hypothesized value (e.g., population mean = 50) lies **outside** the CI, we can reject that hypothesis at the given confidence level.

4. **Decision-Making Tool**

   * Used in medicine (drug effectiveness), business (customer satisfaction surveys), politics (polling), etc.
   * Example: "We are 95% confident that the average customer satisfaction score is between 7.2 and 8.1."

## **Example**

A sample of 100 students has average height $\bar{X} = 165$ cm, with σ = 10 cm.
Find a **95% confidence interval** for population mean.

$$
CI = 165 \pm 1.96 \cdot \frac{10}{\sqrt{100}}
$$

$$
CI = 165 \pm 1.96 \cdot 1
$$

$$
CI = (163.04, 166.96)
$$

So, we are 95% confident that the **true average height** of all students lies between **163.0 and 167.0 cm**.



## 8. What is the concept of expected value in a probability distribution?


## **Expected Value (EV)**

The **expected value** of a random variable is the **long-run average** outcome of a probability distribution after many repetitions of an experiment.

It represents the "center of gravity" of the distribution — the value we expect **on average**.


### **1. For a Discrete Random Variable**

If $X$ takes values $x_1, x_2, \dots, x_n$ with probabilities $P(X=x_i) = p_i$, then:

$$
E[X] = \sum_{i=1}^{n} x_i \cdot p_i
$$


### **2. For a Continuous Random Variable**

If $X$ has a probability density function (PDF) $f(x)$, then:

$$
E[X] = \int_{-\infty}^{\infty} x \cdot f(x) \, dx
$$


## **Examples**

### **Example 1 (Discrete – Dice Roll)**

Roll a fair die (values 1 to 6, each with probability $1/6$):

$$
E[X] = \sum_{i=1}^{6} x \cdot \frac{1}{6} = \frac{1+2+3+4+5+6}{6} = 3.5
$$

 The expected value is **3.5**. (Of course, we never roll 3.5, but it’s the long-run average).



### **Example 2 (Continuous – Uniform Distribution)**

Suppose $X$ is uniformly distributed between 0 and 10.

$$
E[X] = \int_0^{10} x \cdot \frac{1}{10} \, dx
= \frac{1}{10} \cdot \frac{x^2}{2}\Big|_0^{10}
= \frac{100}{20} = 5
$$

 The expected value is **5**, which is the midpoint.


## **Why is Expected Value Important?**

1. **Summarizes distributions** → gives a single representative number.
2. **Decision-making** → used in economics, insurance, gambling, risk analysis.
3. **Foundation for variance** → variance measures how far values deviate from the expected value.
4. **Applied in machine learning/statistics** → expected loss, expected accuracy, etc.



## 9. Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate 1000 random numbers from Normal Distribution (mean=50, std=5)
data = np.random.normal(loc=50, scale=5, size=1000)

# Step 2: Compute mean and standard deviation
mean_val = np.mean(data)
std_val = np.std(data)

print("Sample Mean:", mean_val)
print("Sample Standard Deviation:", std_val)

# Step 3: Plot histogram
plt.figure(figsize=(8,5))
plt.hist(data, bins=30, density=True, color='skyblue', edgecolor='black')
plt.title("Histogram of Normally Distributed Data (μ=50, σ=5)")
plt.xlabel("Value")
plt.ylabel("Frequency Density")
plt.grid(alpha=0.3)
plt.show()
```

## 10. You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend.
## daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
## ● Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval.
## ● Write the Python code to compute the mean sales and its confidence interval.


## **Step 1: Applying Central Limit Theorem (CLT)**

* We have **daily sales data** (sample size $n=20$).
* Population mean is **unknown**, but by CLT, the **sampling distribution of the sample mean** is approximately normal when $n$ is reasonably large (≥30).
* For smaller samples like $n=20$, we typically use the **t-distribution** instead of z, but the logic is the same.

### **95% Confidence Interval Formula**

$$
CI = \bar{X} \pm t_{\alpha/2,\,n-1} \cdot \frac{s}{\sqrt{n}}
$$

Where:

* $\bar{X}$ = sample mean
* $s$ = sample standard deviation
* $n$ = sample size
* $t_{\alpha/2, n-1}$ = critical t-value with $n-1$ degrees of freedom for 95% confidence


```python
import numpy as np
import scipy.stats as stats

# Given daily sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Convert to numpy array
data = np.array(daily_sales)

# Sample statistics
n = len(data)
mean_val = np.mean(data)
std_val = np.std(data, ddof=1)  # sample standard deviation (ddof=1)

# t critical value for 95% CI
t_crit = stats.t.ppf(0.975, df=n-1)

# Margin of error
margin_of_error = t_crit * (std_val / np.sqrt(n))

# Confidence interval
ci_lower = mean_val - margin_of_error
ci_upper = mean_val + margin_of_error

print("Sample Mean:", mean_val)
print("95% Confidence Interval: ({:.2f}, {:.2f})".format(ci_lower, ci_upper))
```
