## Question 1: What is hypothesis testing in statistics?**

**Answer:**
Hypothesis testing is a **statistical method used to make decisions or draw conclusions about a population based on sample data**. It helps determine whether the observed results are due to chance or if they provide enough evidence to support a specific claim (hypothesis).

* In hypothesis testing, we start with two assumptions:

  1. **Null Hypothesis (H₀):** Assumes there is no effect or no difference. (e.g., "The new teaching method has no effect on student performance.")
  2. **Alternative Hypothesis (H₁ or Ha):** Assumes there is an effect or a difference. (e.g., "The new teaching method improves student performance.")

* Steps in hypothesis testing:

  1. State the null and alternative hypotheses.
  2. Select a significance level (α), usually 0.05.
  3. Collect sample data and calculate the test statistic (e.g., t, z, χ²).
  4. Compare the test statistic with the critical value or p-value.
  5. Make a decision:

     * **Reject H₀** if evidence supports H₁.
     * **Fail to reject H₀** if evidence is insufficient.

* **Example:**
  Suppose a company claims their bulb lasts 1000 hours. A sample of 50 bulbs shows an average life of 980 hours. Hypothesis testing can check if the difference is statistically significant or just due to random variation.



## Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?**

**Answer:**

In **statistics**, hypothesis testing is used to make decisions about population parameters based on sample data. The process always involves two competing statements:



### **1. Null Hypothesis (H₀)**

* The **null hypothesis** is a statement that there is **no effect, no difference, or no relationship** in the population.
* It represents the **default assumption** and is considered true until evidence suggests otherwise.
* The null hypothesis is what you **test directly** using sample data.

**Example:**

* A new teaching method does not improve students’ exam scores.

  $$
  H₀: \text{Mean score with new method} = \text{Mean score with old method}
  $$

**Purpose of H₀:**

* Provides a baseline to compare against.
* Helps control the risk of making **Type I errors** (rejecting a true H₀).



### **2. Alternative Hypothesis (H₁ or Ha)**

* The **alternative hypothesis** is a statement that there **is an effect, a difference, or a relationship**.
* It is what the researcher wants to **provide evidence for**.

**Example:**

* A new teaching method improves students’ exam scores.

  $$
  H₁: \text{Mean score with new method} > \text{Mean score with old method}
  $$

**Purpose of H₁:**

* Represents the research claim.
* Accepted **only if the data provides sufficient evidence** against H₀.



### **Summary:**

* **H₀ (Null Hypothesis):** Assumes no change, no effect. It is the “default” claim.
* **H₁ (Alternative Hypothesis):** Assumes change, effect, or difference. It is what we try to support.
* The goal of hypothesis testing is to use **sample evidence** to decide whether to **reject H₀** in favor of H₁.



## Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.**

**Answer:**

* The **significance level (α)** is the **probability of rejecting the null hypothesis (H₀) when it is actually true**.
* It represents the researcher’s tolerance for making a **Type I error** (false positive).
* Common values of α are **0.05 (5%)**, **0.01 (1%)**, or **0.10 (10%)**.

**Role in hypothesis testing:**

1. **Set the threshold for decision-making**

   * Before conducting the test, we fix α (e.g., 0.05).
   * This means we are willing to accept a 5% risk of wrongly rejecting H₀.

2. **Compare with p-value**

   * After performing the test, we calculate the **p-value** (the probability of observing the sample result if H₀ is true).
   * If **p-value ≤ α**, reject H₀ → evidence supports H₁.
   * If **p-value > α**, fail to reject H₀ → evidence not strong enough.

**Example:**
Suppose a company claims a bulb lasts 1000 hours.

* α = 0.05 → We accept a 5% chance of mistakenly rejecting the claim.
* If test result gives p = 0.03 → Since 0.03 < 0.05, we **reject H₀** and conclude the bulb life is significantly different.

**In short:** The significance level sets the **decision boundary** for hypothesis testing. It controls the risk of a false conclusion and determines whether evidence against H₀ is strong enough.



## Question 4: What are Type I and Type II errors? Give examples of each.**

**Answer:**

In hypothesis testing, **errors** occur because decisions are made on the basis of sample data, not the entire population.

### 1. **Type I Error (False Positive)**

* Occurs when we **reject the null hypothesis (H₀) even though it is true**.
* Probability of committing this error is the **significance level (α)**.
* **Example:**

  * A company claims a medicine does **not** cause side effects (H₀).
  * A test incorrectly concludes that the medicine **does cause side effects** (rejects H₀).
  * In reality, the medicine is safe, but we falsely rejected H₀.



### 2. **Type II Error (False Negative)**

* Occurs when we **fail to reject the null hypothesis (H₀) even though it is false**.
* Probability of committing this error is denoted by **β**.
* **Example:**

  * A teacher introduces a new teaching method that actually improves student performance (H₁ is true).
  * But the test concludes there is **no improvement** (fails to reject H₀).
  * In reality, the method is effective, but we missed detecting it.







## Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.**

**Answer:**

The **Z-test** and **T-test** are both statistical tests used in hypothesis testing to determine whether there is a significant difference between sample data and population parameters, or between two samples. However, they differ in assumptions, sample size requirements, and the type of distribution they rely on.



### **1. Z-test**

* The Z-test is used when:

  1. The **population standard deviation (σ)** is known.
  2. The **sample size is large (n > 30)**, so that the sampling distribution approaches normality due to the Central Limit Theorem.
  3. The data follows a **normal distribution** (or approximately normal for large samples).

* It is based on the **standard normal distribution (Z-distribution)**, which has a mean of 0 and standard deviation of 1.

* **Formula:**

  $$
  Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
  $$

  Where:

  * $\bar{X}$ = sample mean
  * $\mu$ = population mean
  * $\sigma$ = population standard deviation
  * $n$ = sample size

* **Example:**
  Suppose the average IQ of a population is 100 with σ = 15. If we take a random sample of 100 people and find the mean IQ is 102, a Z-test can check whether this difference is significant.



### **2. T-test**

* The T-test is used when:

  1. The **population standard deviation (σ) is unknown**.
  2. The **sample size is small (n < 30)**.
  3. The data is approximately **normally distributed**.

* Instead of σ, the **sample standard deviation (s)** is used, and the test follows a **t-distribution**.

* The **t-distribution** looks similar to the normal distribution but has **heavier tails**, which accounts for extra uncertainty when dealing with small samples. As the sample size increases, the t-distribution approaches the normal distribution.

* **Formula:**

  $$
  t = \frac{\bar{X} - \mu}{s / \sqrt{n}}
  $$

  Where:

  * $s$ = sample standard deviation (used instead of σ)

* **Types of T-tests:**

  1. **One-sample T-test:** Checks if the sample mean differs from the population mean.
  2. **Independent two-sample T-test:** Compares the means of two independent groups.
  3. **Paired sample T-test:** Compares means from the same group at two different times or under two different conditions.

* **Example:**
  Suppose a teacher wants to know if the average marks of 20 students differ from 50. Since σ is not known and sample size is small, a T-test is appropriate.



### **Key Difference in Use:**

* **Z-test:** Best for large samples and when σ is known.
* **T-test:** Best for small samples and when σ is unknown.



## Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10   # number of trials
p = 0.5  # probability of success
size = 1000  # number of samples

# Generate binomial distribution
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(n+2), edgecolor='black', alpha=0.7)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

## Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
## sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

Answer:

We will test the hypothesis:

Null Hypothesis (H₀): The population mean is equal to 50.

Alternative Hypothesis (H₁): The population mean is not equal to 50.

Let’s use Z-statistics for this dataset.

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Parameters
population_mean = 50  # Hypothesized mean
alpha = 0.05          # Significance level

# Sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # sample standard deviation
n = len(sample_data)

# Standard error
se = sample_std / np.sqrt(n)

# Z-statistic
z_score = (sample_mean - population_mean) / se

# p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Output results
print(f"Sample Mean: {sample_mean:.3f}")
print(f"Sample Std Dev: {sample_std:.3f}")
print(f"Z-Score: {z_score:.3f}")
print(f"P-Value: {p_value:.4f}")

# Decision
if p_value < alpha:
    print("Reject the Null Hypothesis (H₀).")
else:
    print("Fail to Reject the Null Hypothesis (H₀).")
```

## Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data from a normal distribution
np.random.seed(42)   # for reproducibility
data = np.random.normal(loc=50, scale=5, size=100)  # mean=50, std=5, sample size=100

# Step 2: Calculate sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation
n = len(data)
se = sample_std / np.sqrt(n)

# Step 3: Calculate 95% confidence interval
confidence_level = 0.95
z_critical = stats.norm.ppf((1 + confidence_level) / 2)  # Z value for 95% CI
margin_of_error = z_critical * se
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Step 4: Print results
print(f"Sample Mean: {sample_mean:.3f}")
print(f"95% Confidence Interval: ({lower_bound:.3f}, {upper_bound:.3f})")

# Step 5: Plot the data
plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {sample_mean:.2f}")
plt.axvline(lower_bound, color='green', linestyle='dotted', linewidth=2, label=f"Lower 95% CI = {lower_bound:.2f}")
plt.axvline(upper_bound, color='green', linestyle='dotted', linewidth=2, label=f"Upper 95% CI = {upper_bound:.2f}")
plt.title("Normal Distribution with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

## Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores
def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample standard deviation
    z_scores = (data - mean) / std
    return z_scores

# Sample dataset
data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5]

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Print Z-scores
print("Z-scores:", np.round(z_scores, 3))

# Visualize Z-scores using histogram
plt.hist(z_scores, bins=10, edgecolor='black', alpha=0.7)
plt.title("Histogram of Z-scores")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```


### **Explanation:**

* A **Z-score** measures how many **standard deviations a data point is from the mean**.

  * Formula:

    $$
    Z = \frac{X - \bar{X}}{s}
    $$

    Where:

    * $X$ = data point
    * $\bar{X}$ = sample mean
    * $s$ = sample standard deviation

* **Interpretation:**

  * **Z = 0:** Data point equals the mean.
  * **Z > 0:** Data point is above the mean.
  * **Z < 0:** Data point is below the mean.
  * **|Z| = 1:** Data point is 1 standard deviation away from the mean.

* **Histogram:** Visualizes the distribution of standardized values, showing how data points spread around the mean in terms of standard deviations.

This helps in identifying **outliers** and comparing values across different datasets.
