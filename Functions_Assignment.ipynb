### 3. What are the different ways to define and call a function in Python?

-> A function is a block of reusable code that performs a specific task. You can define functions in different ways based on the purpose and data handling. Once defined, functions can be called or invoked to execute the code.

Here are the different ways to define and call functions in Python:

1. Simple Function (No Parameters):
This is the most basic function. It performs an action but doesn't accept any input.

2. Function with Parameters:
You can pass values to a function using parameters, which makes the function dynamic.

3. Function with Default Parameters:
Parameters can have default values, which are used if no argument is passed during the function call.

4. Function with Return Statement:
A function can return a value using the return keyword. This allows you to store or reuse the result.

5. Lambda Function (Anonymous Function):
Lambda functions are small, unnamed functions defined using the lambda keyword. They are useful for short, simple operations.

6. Recursive Function:
A recursive function is a function that calls itself. It is commonly used in problems that can be broken into smaller sub-problems (like factorial, Fibonacci, etc.).

```python
# 1. Simple Function (No Parameters)
def greet():
    print("Hello!")

greet()

# 2. Function with Parameters
def greet_person(name):
    print(f"Hello, {name}!")

greet_person("Suraj")

# 3. Function with Default Parameters
def greet_default(name="Guest"):
    print(f"Hello, {name}!")

greet_default()          # Uses default value
greet_default("Riya")    # Uses provided value

# 4. Function with Return Statement
def add(a, b):
    return a + b

result = add(10, 5)
print("Sum:", result)

# 5. Lambda Function
square = lambda x: x * x
print("Square:", square(6))

# 6. Recursive Function
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)

print("Factorial of 5:", factorial(5))
```

### 5. What are iterators in Python and how do they differ from iterables?

-> Iterators and iterables are used in loops (like for loops), but they are not the same:

* Iterable:

An iterable is any object that can be looped over.

Examples: list, tuple, string, dict, set.

It has a built-in method called __iter__() which returns an iterator.

* Iterator:

An iterator is an object that keeps track of its position while looping.

It has two main methods:

* __iter__() → returns the iterator object itself.

* __next__() → returns the next value in the sequence.

Once all items are exhausted, __next__() raises a StopIteration exception.

```python
# Example

# A list is an iterable
my_list = [10, 20, 30]

# Convert iterable to iterator
my_iterator = iter(my_list)

# Get values using the iterator
print(next(my_iterator))  # Output: 10
print(next(my_iterator))  # Output: 20
print(next(my_iterator))  # Output: 30
```

### 6. Explain the concept of generators in Python and how they are defined.

-> A generator in Python is a special type of iterator that allows you to generate values on the fly without storing them all in memory. It is used to produce a sequence of values, but unlike lists, it doesn’t store them all at once — it yields one value at a time.

#### How are Generators Defined?

* Generators are defined like normal functions but use the yield keyword instead of return.

* When a generator function is called, it returns a generator object, but does not run the function immediately.

* Each call to next() executes the function up to the next yield statement.

#### Why Use Generators?

* Memory efficient – they don’t store all values at once.

* Useful for large datasets or infinite sequences

* Can be used in loops, just like iterators.

```python
#Example

def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

# Create generator object
counter = count_up_to(3)

# Get values one by one
print(next(counter))
print(next(counter))
print(next(counter))
```

### 7. What are the advantages of using generators over regular functions?
-> Generators provide a powerful way to create iterators with a simple and clean syntax. They are especially useful when working with large data sets or streams of data. Compared to regular functions that return full lists or other collections, generators offer several advantages:

1. Memory Efficiency:

Generators produce items one at a time using the yield statement. They don’t store the entire sequence in memory, unlike regular functions that return a list. This makes them ideal for large or even infinite data streams.

2. Lazy Evaluation:

Generators use lazy evaluation, meaning they only compute values when needed. This avoids unnecessary calculations and speeds up execution when only part of the data is required.

3. Cleaner and Simpler Code:

Generators allow you to write stateful logic in a much cleaner way using yield. You don’t have to manually track loop counters or states like in regular functions.

4. Infinite Sequence Support:

Regular functions must generate all values before returning them. But generators can generate an infinite series (like Fibonacci numbers, or counting forever) without using up memory.

5. Better Performance in Loops:

Since generators return values one at a time, loops using generators are faster and use less memory compared to loops using large lists.



```python
#Example

# Regular function that returns a list
def get_numbers_list(n):
    return [i for i in range(n)]

# Generator function
def get_numbers_gen(n):
    for i in range(n):
        yield i

# Using both
print("Using list:", get_numbers_list(3))

gen = get_numbers_gen(3)
print("Using generator:")
print(next(gen))
print(next(gen))
print(next(gen))
```

### 8. What is a lambda function in Python and when is it typically used?

-> A lambda function in Python is a small anonymous (unnamed) function that is defined using the lambda keyword instead of def.

Lambda functions are also known as:

* Anonymous functions

* Inline functions

* One-liner functions

#### When is a Lambda Function Typically Used?

* You need a short, throwaway function.
* You’re using functions like map(), filter(), sorted(), or reduce().
* You don’t want to define a full def function for a small task.
* You need a quick function inside GUI callbacks, event handlers, or sorting custom keys.

```python
# Example

# Lambda to add two numbers
add = lambda x, y: x + y
print(add(3, 5))  # Output: 8

# Lambda inside sorted() to sort by second element
pairs = [(1, 4), (2, 1), (3, 2)]
sorted_pairs = sorted(pairs, key=lambda pair: pair[1])
print(sorted_pairs)
```

### 9.  Explain the purpose and usage of the map() function in Python.

-> The map() function in Python is used to apply a function to each item of an iterable (like a list, tuple, etc.) and returns a map object (which is an iterator).

It allows you to transform data without using an explicit loop. This makes your code shorter, cleaner, and more efficient, especially when performing the same operation on every element.

The general syntax is:

map(function, iterable)

The function defines what action to perform.

The iterable is the data you want to apply that function to.

You usually wrap the result with list() or tuple() to view the output.

```python
# Example

numbers = [1, 2, 3, 4]
squared = list(map(lambda x: x * x, numbers))
print(squared)
```

### 10.  What is the difference between map(), reduce(), and filter() functions in Python?

1. map()

Purpose: Transforms each item in an iterable using a given function.

Returns: A new iterable with the transformed values.

Use case: When you want to modify every element.

2. filter()

Purpose: Filters elements from an iterable based on a condition (True/False).

Returns: Only those elements for which the function returns True.

Use case: When you want to remove unwanted elements based on a rule.

3. reduce()

Purpose: Applies a function to the iterable cumulatively, reducing it to a single value.

Returns: A single output.

Use case: When you want to combine all items (like sum, product, etc.).

Note: reduce() is not built-in by default in Python 3 — you need to import it from functools.

```python
# map example

numbers = [1, 2, 3]
squared = list(map(lambda x: x * x, numbers))
print(squared)
# filter example

numbers = [1, 2, 3, 4]
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)

#reduce example

from functools import reduce

numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)
```

```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_of_even(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

# Example usage
nums = [10, 5, 7, 8, 2]
result = sum_of_even(nums)
print("Sum of even numbers:", result)
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

# Example usage
number = 113
if 1 <= number <= 200:
    result = is_prime(number)
    print(f"{number} is prime:", result)
else:
    print("Number is out of range (1 to 200)")
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, n):
        self.n = n        # total terms
        self.count = 0    # current term index
        self.a = 0        # first number
        self.b = 1        # second number

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.count += 1
            self.a, self.b = self.b, self.a + self.b
            return self.b

fib = FibonacciIterator(7)  # Generate 7 terms

for num in fib:
    print(num, end=" ")
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string

def remove_vowels(text):
    vowels = 'aeiouAEIOU'
    result = ''.join(filter(lambda ch: ch not in vowels, text))
    return result

# Example usage
input_str = "Suraj Gupta"
output_str = remove_vowels(input_str)
print("String without vowels:", output_str)
```
